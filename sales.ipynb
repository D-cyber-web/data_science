```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales.csv')
```

```python
df
```

```python
#help(pd.pivot)
```

```python
Quantity = df[['Name','Product','Quantity']]
```

```python
pd.pivot(data=Quantity, index='Name', columns= 'Product',values='Quantity')
```

```python
pd.pivot_table(df,index="Name",aggfunc='sum')
```

```python
pd.pivot_table(df,index="Name",aggfunc='sum',values=['Price','Quantity'])
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'],columns=['Product'],aggfunc=[np.sum],fill_value=0,margins=True)
```
